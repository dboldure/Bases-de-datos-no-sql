## Sensor


Notebook auxiliar para la práctica. Simula la emisión de datos por parte de un sensor. Ponerlo en marcha antes de probar la práctica, para que ya esté emitiendo datos

```python
redisconfPath = "c:/hlocal/"

#redisconfPath = "c:/hlocal/"

SENSORES=3
MINUTOS=80
import sys
sys.path.insert(0, redisconfPath)
import redisconf

import  redis
import time
from random import randint

pool = redis.ConnectionPool(host=redisconf.CONEXION, password=redisconf.PASSWD, 
                port=redisconf.PUERTO,db=0)
r = redis.Redis(connection_pool=pool)
if r.ping():
    # milisegundos
    lapse = 1000
    t = 0
    start = time.time()
    # funcionará durante MINUTOS minutos, luego se para
    while time.time() - start<MINUTOS*60:

        # emitimos datos aleatorios
        sensor = randint(1, SENSORES)
        if randint(1, 6)==1: # error de lectura cada 6 veces
            valor=None
        else:
            valor = (10**sensor)*randint(1, 9)
        print("time: ",t,". Sensor: ",sensor,". Valor: ",valor)
        if valor==None:
            r.delete("value")
            datos = {"sensor":str(sensor),"time":t}
        else:
            datos = {"sensor":str(sensor),"time":t,"value":  valor }
        r.mset(datos)
        time.sleep(lapse/1000)
        t+=1
        
else:
    print("Fallo en la conex. Iniciaste el servidor?")
```
